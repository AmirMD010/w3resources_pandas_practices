# Pandas Advanced Merging and Joining: Question 1 to

1. Write a Pandas program to merge two DataFrames on a single column.

```python
data1 = pd.DataFrame({'ID': [1, 2, 3],'Name': ['Selena', 'Annabel', 'Caeso']})
data2 = pd.DataFrame({'ID': [2, 3, 4],'Age': [25, 30, 22]})

df = pd.merge(data1, data2, on='ID')
df
```

2.Write a Pandas program to perform an outer join on two DataFrames.

```python
df = pd.merge(data1, data2, on ='ID', how='outer')
df
```

3. Write a Pandas program that performs a left join of two DataFrames.

```python
df = pd.merge(data1,data2, on='ID', how='left')
df
```

4. Write a Pandas program that performs a right join of two DataFrames.

```python
df = pd.merge(data1,data2, on='ID', how='right')
df
```

5. Write a Pandas program to merge two DataFrames on multiple columns.

```python
data1 = pd.DataFrame({'ID': [1, 2, 3],'Name': ['Selena', 'Annabel', 'Caeso'],'Age': [25, 30, 22]})
data2 = pd.DataFrame({'ID': [1, 2, 3],'Name': ['Selena', 'Annabel', 'Caeso'],'Salary': [50000, 60000, 70000]})

df = pd.merge(data1,data2, on=['ID','Name'])
df
```

6. Write a Pandas program that merges DataFrames with overlapping column names.

```python
data1 = pd.DataFrame({'ID': [1, 2, 3],'Name': ['Selena', 'Annabel', 'Caeso'],'Value': [100, 200, 300]})
data2 = pd.DataFrame({'ID': [1, 2, 3],'Name': ['Selena', 'Annabel', 'Caeso'],'Value': [400, 500, 600]})

df = pd.merge(data1,data2, on=['ID','Name'], suffixes=(' left',' right'))
df
```

7. Write a Pandas program to merge two DataFrames on their indexes.

```python
data1 = pd.DataFrame({
    'Name': ['Selena', 'Annabel', 'Caeso'],
    'Age': [25, 30, 22]
}, index=[1, 2, 3])

data2 = pd.DataFrame({
    'Salary': [50000, 60000, 70000]
}, index=[1, 2, 3])

df = pd.merge(data1,data2, left_index=True, right_index=True)
df
```

8. Write a Pandas program to merge different column names in DataFrames.

```python
data1 = pd.DataFrame({
    'Employee_ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})

data2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Salary': [50000, 60000, 70000]
})

df = pd.merge(data1,data2, left_on='Employee_ID', right_on='ID')
df
```

9. Write a Pandas program to merge DataFrames with duplicate Keys.

```python
data1 = pd.DataFrame({'ID': [1, 1, 3],'Name': ['Selena', 'Annabel', 'Caeso']})
data1 = pd.DataFrame({'ID': [1,2],'Age': [16, 20]})

df = pd.merge(data1,data2, on='ID')
df
```

10. Write a Pandas program to merge multiple DataFrames on a common column.

```python
data1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})

data2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Age': [25, 30, 22]
})

data3 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Salary': [50000, 60000, 70000]
})

df = pd.merge(pd.merge(data1,data2),data3,on='ID')
df
```

11. Write a Pandas program to merge DataFrames using join() on Index.

```python
data1 = pd.DataFrame({
    'Name': ['Selena', 'Annabel', 'Caeso'],
    'Age': [25, 30, 22]
}, index=[1, 2, 3])

data2 = pd.DataFrame({
    'Salary': [50000, 60000, 70000]
}, index=[1, 2, 3])

df = data1.join(data2)
df
```

12. Write a Pandas program to merge with custom indicator to track source.

```python
data1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})

data2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Age': [25, 30, 22]
})

df = pd.merge(data1,data2, how='outer', indicator=True)
df
```

13. Write a Pandas program to merge two DataFrames using multiple keys and specific join conditions.

```python
data1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Selena', 'Annabel', 'Charlie', 'Caeso'],
    'City': ['NY', 'LA', 'NY', 'LA']
})

data2 = pd.DataFrame({
    'ID': [2, 3, 4, 5],
    'Name': ['Annabel', 'Charlie', 'Caeso', 'Eve'],
    'City': ['LA', 'NY', 'LA', 'NY'],
    'Age': [30, 22, 25, 28]
})

df = pd.merge(data1,data2, on=['ID','City'],how='inner')
df
```

14. Write a Pandas program to merge DataFrames using suffixes for overlapping columns.

```python
df = pd.merge(data1,data2, on=['ID','City'],how='inner', suffixes=('_left', '_right'))
df
```

15. Write a Pandas program to merge DataFrames with custom sorting.

16. Write a Pandas program to merge DataFrames and drop duplicates.

```python
data1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Annabel', 'Selena', 'Caeso']
})

data2 = pd.DataFrame({
    'ID': [2, 3, 1],
    'Name': ['Selena', 'Caeso', 'Annabel'],
    'Age': [30, 22, 25]
})

df = pd.merge(data1,data2, on=['ID','Name']).drop_duplicates()
df
```

17. Write a Pandas program to merge DataFrames with missing data.

```python
data1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']
})

data2 = pd.DataFrame({
    'ID': [1, 3, 4],
    'Age': [25, 22, 28]
})

df = pd.merge(data1,data2, on='ID',how='outer').fillna('Unknown')
df
```

19. Write a Pandas program to merge DataFrames on multiple columns with different names.

```python
data1 = pd.DataFrame({
    'Employee_ID': [1, 2, 3],
    'Name': ['Selena', 'Annabel', 'Caeso']  # Matching names with df2
})

data2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Employee_Name': ['Selena', 'Annabel', 'Caeso'],  # Matching names with df1
    'Salary': [50000, 60000, 70000]
})

df = pd.merge(data1,data2, left_on=['Employee_ID','Name'],right_on=['ID','Employee_Name'])
df
```
